```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ct = nx.star_graph(100)
```

```python
pos = nx.spring_layout(ct, iterations=20)

colors = range(100)

nx.draw(ct, pos, node_color='b', edge_color=colors, width=4, edge_cmap=plt.cm.Blues, with_labels=True)
```
