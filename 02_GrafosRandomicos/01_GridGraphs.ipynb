# Grid Graphs

```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
G = nx.grid_2d_graph(4,4)  # 4x4 grid

pos = nx.spring_layout(G, iterations=100)
```

```python
# plot 1
plt.subplot(221)
nx.draw(G, pos, font_size=8)

# plot 2
plt.subplot(222)
nx.draw(G, pos, node_color='k', node_size=0, with_labels=False)

# plot 3
plt.subplot(223)
nx.draw(G, pos, node_color='g', node_size=100, with_labels=False)

# plot 4
plt.subplot(224)

H = G.to_directed()
nx.draw(H, pos, node_color='b', node_size=100)
```
