# Funções para Gerar Grafos

```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Definindo uma função em python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([
        (1,2),
        (2,3),
        (1,3),
        (1,4)
    ])
    nx.draw(G, node_size=800, node_color='b',with_labels=True)
```

```python
draw_graph()
```

```python
# Definir função Draw Barabase Albert
def draw_barabasi_albert():
    G = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(G)
    
    plt.figure(figsize=(20,15))
    plt.subplot(221)
    nx.draw(G, with_labels=True)
    plt.subplot(222)
    plt.hist(hst, bins=20, color='red')
    plt.show()
```

```python
draw_barabasi_albert()
```

```python
# Nova função
def simple_graph():
    #create our empty graph
    G = nx.Graph()
    
    #add edges
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    #draw the graph
    nx.draw(G, with_labels=True, node_color='y',node_size=800)
```

```python
simple_graph()
```

```python
# Removendo um node
def delete_nodes_in_graph():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G, with_labels=True, node_size=800)
    
    # remove node B
    G.remove_node('B')
    
    nx.draw(G, with_labels=True, node_size=800, node_color='b')
```

```python
delete_nodes_in_graph()
```

```python
# Removendo edges
def delete_specific_edge_in_graph():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G, with_labels=True, node_size=800)
    
    # remove node B
    G.remove_edge('A','B')
    
    nx.draw(G, with_labels=True, node_size=800, node_color='b')
```

```python
delete_specific_edge_in_graph()
```

```python
# Customizar caracteristicas dos nodes
def custom_node_size_and_node_color():
    G = nx.Graph()
    
    # add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G)
    plt.show()
    
    #define node list
    nodelist = ['A','B','C']
    
    # define the color list
    colorlist = ['c','m','y']
    
    # define the node size list
    sizelist = [1000, 2000, 3000]
    
    nx.draw(G,
           with_labels=True,
           node_color=colorlist,
           node_size=sizelist,
           nodelist=nodelist)
```

```python
custom_node_size_and_node_color()
```
