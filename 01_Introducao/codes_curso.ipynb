### Clustering Coefficient

c_u = $\frac{2 T(u)}{deg(u)(deg(u)-1)}$,

where `T(u)` is the number of triangles through node `u` and
`deg(u)` is the degree of `u`.

```python
CC = nx.complete_graph(5)
nx.draw_networkx(CC,node_color='cyan')
```

```python
print(nx.clustering(CC))
```

### Clustering
Compute the average clustering coefficient for the graph G.

The clustering coefficient for the graph is the average

C = $\frac{1}{n}\sum_{v \in G} c_v$,
   
where 'n' is the number of nodes in `G`.

```python
print(nx.average_clustering(CC))
```

## Bipartite

Bipartite graphs are graphs that have two (bi-) partitions (-partite) of nodes. Nodes within each partition are not allowed to be connected to one another; rather, they can only be connected to nodes in the other partition.

Bipartite graphs can be useful for modelling relations between two sets of entities. We will explore the construction and analysis of bipartite graphs here.

```python
import networkx as nx
from networkx.algorithms import bipartite
```

```python
B = nx.Graph()
```

```python
#define cities
cities = ['Beijing','Xian','Vancouver','San Francisco','Austin','Boston']

# define people
people = [ 'Theo ', 'Student']

# adding edges of cities and people
B.add_nodes_from(cities, bipartite='cities')
B.add_nodes_from(people,bipartite='people')

# define edges
edges = [('Theo','Vancouver'),('Student','Xian'),('Theo','San Francisco'),('Student','Boston'),('Theo','Boston'),('Student','Beijing')]

B.add_edges_from(edges)

nx.draw(B,with_labels=True, node_color='cyan',node_size=750)
```

Note that if the number of nodes in the graph gets really large, node-link diagrams can begin to look like massive hairballs. This is undesirable for graph visualization.
Instead, we can use a matrix to represent them. The nodes are on the x- and y- axes, and a filled square represent an edge between the nodes. This is done by using the nx.to_numpy_matrix(G) function.
We then use matplotlib's pcolor(numpy_array) function to plot. Because pcolor cannot take in numpy matrices, we will cast the matrix as an array of arrays, and then get pcolor to plot it.

```python
plt.figure(figsize=(10,7))

matrix = nx.to_numpy_matrix(social)

plt.pcolor(np.array(matrix))

plt.axes().set_aspect('equal')

plt.xlim(min(social.nodes()), max(social.nodes()))
plt.ylim(min(social.nodes()), max(social.nodes()))

plt.title("Adjacency Matrix")

plt.show()
```

Let's try another visualization, the Circos plot. We can order the nodes in the Circos plot according to the node ID, but any other ordering is possible as well. Edges are drawn between two nodes. Credit goes to Justin Zabilansky (MIT) for the implementation.

```python
from circos import CircosPlot
```

```python
fig = plt.figure(figsize=(10,7))

ax = fig.add_subplot(111,axisbg='c')

nodes = sorted(social.nodes())
edges = social.edges()

c = CircosPlot(nodes,edges,radius=10,nodecolor='black',ax=ax,figsize=(8,8))

c.draw()
```

It's pretty obvious in this visualization that there are nodes, such as node 5 and 18, that are not connected to any other node via an edge. There are other nodes, like node number 19, which is highly connected to other nodes. Finally, let's try hive plots for the network. Two groups (male and female), and then edges drawn between them.

```python
from hiveplot import HivePlot
```

```python
nodes = dict()

nodes['male'] = [n for n,d in social.nodes(data=True) if d['sex'] == 'Male']
nodes['female'] = [n for n,d in social.nodes(data=True) if d['sex'] == 'Female']

edges = dict()

edges['group1']= social.edges(data = True)

nodes_cmap = dict()
nodes_cmap['male'] = 'blue'
nodes_cmap['female'] = 'red'

edges_cmap = dict()
edges_cmap['group1'] = 'black'

h = HivePlot(nodes, edges, nodes_cmap, edges_cmap)

h.draw()

 
```

### Let us say a Graph G = {(A,B), (B,C), (C,A)}, where each tuple in the set denotes an edge

How to create and visualize this graph G?

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#define our function called simple -graph

def simple_graph():
    #create our empty graph
    G = nx.Graph()
    
    #add edges
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    #draw the graph
    nx.draw(G,with_labels=True,node_color='y',node_size=800)
    
```

```python
#call simple_graph function

simple_graph()
```

## How to delete nodes from a graph

```python
def delete_nodes_in_graph():
    G = nx.Graph()
    
    #add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G, with_labels=True,node_color='y',node_size=800)
    
    #remove node B from the graph
    G.remove_node('B')
    
    nx.draw(G,with_labels=True,node_color='c',node_size=800)
    
```

```python
delete_nodes_in_graph()
```

## Delete specific edges {(A,B), (A,C)} from G

```python
def delete_specific_edge_in_graph():
    G = nx.Graph()
    
    #add 3 edges to the graph
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G,with_labels=True)
    plt.show()
    
    #remove 2 edges from the graph
    G.remove_edge('A','B')
    G.remove_edge('A','C')
    
    nx.draw(G,with_labels=True)
    plt.show()
    
    
    
    
```

```python
delete_specific_edge_in_graph()
```

## Customize edge colors in the graph

```python
def custom_edge_colors():
    G = nx.Graph()
    
    #ADD 3 edges
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G)
    plt.show()
    
    #create list of edge colors
    edgecol = ['c','y','m']
    
    #create node sizes
    nodesize = [ 1000,2000,3000]
    
    #create the edgelist
    edgelst = [('A','B'),('B','C'),('C','A')]
    
    nx.draw(G,edgelist=edgelst,edge_color=edgecol,with_labels=True,node_size=nodesize)
    plt.show()
    
```

```python
#call the function
custom_edge_colors()
```
