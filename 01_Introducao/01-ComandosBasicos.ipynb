```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

## Comandos b√°sicos

```python
# Criar um Grafo vasio sem nodes e edges
G = nx.Graph()

# Adicionar nodes
G.add_nodes_from([2,3,4,5])

# Visualizar
nx.draw_networkx(G, node_color='blue', node_size=700)
```

```python
# Adicionar edges no grafo
G.add_edge(1,2)

nx.draw_networkx(G)
```

```python
# Adicionar edges a partir de uma tupla
e = (2,3)
G.add_edge(*e)

nx.draw_networkx(G)
```

```python
# Adicionar lista de edges
list_e = [(1,2),(1,3),(3,4),(4,5)]
G.add_edges_from(list_e)

nx.draw_networkx(G)
```

```python
# N√∫mero de nodes
print("N nodes: ", G.number_of_nodes())

# N√∫mero de edges
print("N edges: ", G.number_of_edges())
```

```python
# Propriedades do Grafo
print("Nodes:\n",G.nodes())
print("Edges:\n", G.edges())
print("Nodes 1 est√° conectado com:\n", [x for x in G.neighbors(1)])
```

```python
print("Numero de edges por node\n",G.degree())
print("H√° liga√ß√£o entre node 2 e 4?\n", G.has_edge(2,4))
print("H√° liga√ß√£o entre node 1 e 3?\n", G.has_edge(1,3))
print("Existe o node 5 no grafo?\n", G.has_node(5))
```

## Manipula√ß√£o dos Nodes

```python
# Criar Grafo
H = nx.Graph()

# Criar lista de edges e nodes
edgelist = [(0,1),(1,2),(2,3),(0,3),(1,3)]

# Adicionar edges e nodes ao Grafo
H.add_edges_from(edgelist)

# Desenhar grafo
nx.draw(H, with_labels=True, node_color="green", node_size=700)
```

```python
# Remover node 3
H.remove_node(3)

# Desenhar grafo
nx.draw(H, with_labels=True, node_color="green", node_size=700)
```

```python
# Remover uma lista de nodes
list_remove_nodes = [0,1]
H.remove_nodes_from(list_remove_nodes)

# Desenhar grafo
nx.draw(H, with_labels=True, node_color="green", node_size=700)
```

```python
# Novo Grafo
H = nx.Graph()

# adicionar nodes e edges
H.add_edge(1,2, weight=5.5)
H.add_edges_from([(3,4),(4,5)], color='blue')
H.add_edges_from([
    (1,2,{'color':'blue'}),
    (2,3,{'weight':8})
])
nx.draw_networkx(H,with_labels=True,node_color='green',node_size=900)
```

```python
H[2][3]['weight'] = 4.7
```

```python
# Visualizar propriedades dos edges
H.edges(data=True)
```

```python
# Grafo de dist√¢ncia entre cidades
cities = nx.Graph()

cities.add_edge('S√£o Paulo','Rio de Janeiro',distance = 0.4)
cities.add_edge('Belo Horizonte','S√£o Paulo',distance =0.6)
cities.add_edge('Rio de Janeiro','Vit√≥ria',distance =0.3)
cities.add_edge('Belo Horizonte','Vit√≥ria',distance =0.5)

cities.edges(data=True)
```
