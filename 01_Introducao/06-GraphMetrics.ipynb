```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(10,7))
N = nx.path_graph(5, create_using=nx.DiGraph())

nx.draw_networkx(N, node_color='cyan',node_size=700)
```

```python
# Encontrar o caminha mais curto (Shortest path) entre dois nodes
print(nx.shortest_path(N, source=0, target=4))
```

# Clustering Coefficient

c_u = $\frac{2 T(u)}{deg(u)(deg(u)-1)}$,

where `T(u)` is the number of triangles through node `u` and
`deg(u)` is the degree of `u`.

## NetworkX
https://networkx.github.io/documentation/latest/reference/algorithms.clustering.html

https://networkx.github.io/documentation/latest/reference/generated/networkx.algorithms.cluster.average_clustering.html

## Wikipedia
https://en.wikipedia.org/wiki/Clustering_coefficient

```python
CC = nx.complete_graph(5)
nx.draw_networkx(CC, node_color='cyan')
```

```python
print(nx.clustering(CC))
```

### Clustering
Compute the average clustering coefficient for the graph G.

The clustering coefficient for the graph is the average

C = $\frac{1}{n}\sum_{v \in G} c_v$,
   
where 'n' is the number of nodes in `G`.

```python
print(nx.average_clustering(CC))
```
