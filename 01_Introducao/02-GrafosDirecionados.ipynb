# Grafos Direcionados

```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Criar novo grafo
b = nx.DiGraph([
    ('a','b'),
    ('b','c'),
    ('c','a')
])
```

```python
# adicionar edges direcionados
b.add_edge = ('a','b')
b.add_edges_from([
    ('d','c'),
    ('d','b')
])
```

```python
# Visualizar
nx.draw_networkx(b, with_labels=True)
```

```python
# visualizar o out-degrees do grafo direcionado
print(b.out_degree())
```

```python
# visualizar o in-degrees do grafo direcionado
print(b.in_degree())
```

```python
# visualizar outras propriedades
print("In-edges:\n", b.in_edges())
print("Out-edges:\n", b.out_edges())
```

```python
# Novo Grafo Direcionado
dg = nx.DiGraph()

dg.add_edges_from([
    (1,2),
    (1,3),
    (2,4),
    (2,5),
    (2,6),
    (3,7),
    (7,8)
])
```

```python
nx.draw(dg, node_color='#9999FF', with_labels=True)
```

```python
# Metricas
print("Grau de Centralidade:\n", nx.degree_centrality(dg))
```
