```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Desenhar um grafo completo
plt.figure(figsize=(10,7))

cg = nx.complete_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(cg)
```

```python
print(cg.in_degree())
```

```python
# Desenhar Grafo denomidado Path-Graph
plt.figure(figsize=(10,7))

pg = nx.path_graph(5, create_using=nx.DiGraph())
nx.draw_networkx(pg)
```

```python
# Desenhar um GNR-Graph, como grafos de redes sociais(facebook, twitter)
plt.figure(figsize=(10,7))

gnr = nx.gnr_graph(100, 1, create_using=nx.DiGraph())
nx.draw_networkx(gnr)
```

```python
print(gnr.in_degree())
```

```python
# Complete-bipartite-graph
plt.figure(figsize=(10,7))

cbg = nx.complete_bipartite_graph(20,15)
nx.draw_networkx(cbg)
```

```python
print(cbg.size())
```

```python
# Grid-Graph
plt.figure(figsize=(10,7))

gg = nx.grid_graph(dim=[5,5])
nx.draw_networkx(gg, node_size=500)
```

```python
print(gg.order())
```

```python
print(gg.size())
```
