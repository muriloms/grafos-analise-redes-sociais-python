```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

# Karate Club Graph
O clube de karatê de Zachary é uma rede social de um clube universitário de karatê, descrito no artigo "Um modelo de fluxo de informações para conflitos e fissão em pequenos grupos", de Wayne W. Zachary.

```python
k = nx.karate_club_graph()

nx.draw(k, node_color="g", node_size=600, with_labels=True)
```
