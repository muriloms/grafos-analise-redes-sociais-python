```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

# Grafo de Petersen
No campo da matemática da teoria dos grafos o grafo de Petersen é um grafo não-orientado com 10 vértices e 15 arestas. É um pequeno grafo que serve como um exemplo útil e contra-exemplo para muitos problemas em teoria dos grafos.

```python
p = nx.petersen_graph()

nx.draw(p, node_color="cyan", node_size=600, with_labels=True)
```

# Grafo de Tutte
No campo matemático da teoria dos grafos, o gráfico de Tutte é um gráfico tridimensional com 46 vértices e 69 arestas com o nome de W. T. Tutte. Possui número cromático 3, índice cromático 3, perímetro 4 e diâmetro 8. O gráfico Tutte é um gráfico poliédrico cúbico, mas não é hamiltoniano.

```python
t = nx.tutte_graph()

nx.draw(t, node_color="b", node_size=600, with_labels=True)
```

# Grafo de labirinto sedgewick

```python
a = nx.sedgewick_maze_graph()

nx.draw(a, node_color="c", node_size=600, with_labels=True)
```
