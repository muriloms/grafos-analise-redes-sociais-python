```python
# Carregar pacotes
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

# Barbell Graph
o gráfico n-barbell é um tipo especial de gráfico não direcionado que consiste em duas panelinhas n-vértices não sobrepostas, juntamente com uma única aresta que tem um ponto final em cada clique

```python
plt.figure(figsize=(10,7))
           
barbel = nx.barbell_graph(5,5)

nx.draw(barbel, node_color="red", node_size=600, with_labels=True)
```

# Lollipop Graph
o gráfico pirulito é um tipo especial de gráfico que consiste em um gráfico completo em m vértices e um gráfico de caminho em n vértices, conectado a uma ponte.

```python
plt.figure(figsize=(10,7))

lolli = nx.lollipop_graph(5, 10)

nx.draw(lolli, node_color="b", node_size=600, with_labels=True)
```

# Binomial Graph

```python
binom = nx.binomial_graph(50, 0.15)

nx.draw(binom, node_color="g", node_size=600, with_labels=True)
```

# Chordal Cycle Graph

```python
chordal = nx.chordal_cycle_graph(20)

nx.draw(chordal, node_color="#A0CBE2", node_size=600, with_labels=True)
```

```python
k_3_5 = nx.complete_bipartite_graph(10,15)

nx.draw(k_3_5, node_size=600, with_labels=True)
```
